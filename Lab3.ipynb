```python
def has_duplicate(lst):
    lst.sort()
    var=0
    for i in range(len(lst)):
        if lst[i]== var:
            return True
        else:
            var=lst[i]            
    return False
lst1 = [10,20,40,30,40,50]
print(has_duplicate(lst1))
```
